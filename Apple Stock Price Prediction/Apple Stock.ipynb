```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("Desktop/AAPL.csv")
data.head()
```

```python
data.columns
```

```python
data = data.drop(['Unnamed: 0','symbol','splitFactor','volume','divCash','date','adjVolume'],axis=1)
```

```python
X = data.drop('high',axis=1)
y = data['high']
```

```python
sns.barplot(data['date'],data['high'])
```

```python
sns.pairplot(data=data)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pre = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_test,y_pre))
```

```python
nums = [i for i in range(1,len(y_test)+1)]
```

```python
plt.scatter(nums,y_pre,color="blue")
plt.scatter(nums,y_test,color="red")
plt.show()
```
